# **Q1. Explain the different types of data (qualitative and quantitative)...**
 Data is classified into qualitative (non-numeric) and quantitative (numeric).
Qualitative data includes:

Nominal: Categories without order (e.g., gender, colors).

Ordinal: Categories with order (e.g., rankings).

Quantitative data includes:

Interval: Numeric data without a true zero (e.g., temperature).

Ratio: Numeric data with a true zero (e.g., weight, height).

Understanding data types helps in choosing appropriate statistical methods for analysis and interpretation.

#**Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.**
Central tendency locates the center of data.

Mean: Arithmetic average, used in symmetric data.

Median: Middle value, better for skewed data.

Mode: Most frequent value, best for categorical data.

Example: In salaries, median is preferred as mean gets affected by outliers. Mean is used in marks calculation; mode is used in identifying most sold products.

# **Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**
Dispersion measures how much data varies from the center.

Variance: Average of squared deviations from the mean.

Standard Deviation: Square root of variance, easier to interpret.

Low dispersion = more consistency; high dispersion = more variability.
Example: Consistent student scores have low standard deviation; variable scores indicate inconsistency. These help in comparing datasets’ spread.

# **Q4. What is a box plot, and what can it tell you about the distribution of data?**
A box plot visually shows data distribution using five-number summary: minimum, Q1, median, Q3, maximum.
It highlights:

Spread

Central value

Outliers

Example: In exam results, a box plot shows performance range and outlier scores. It helps compare multiple datasets and detect skewness quickly.

# **Q5. Discuss the role of random sampling in making inferences about populations.**
Random sampling ensures each population member has equal selection chance.
It reduces bias and provides reliable, generalizable results.
Example: Surveying random voters gives better predictions than targeting one group. Random sampling underpins statistical inference like hypothesis testing and confidence intervals, ensuring fair representation of the population.

# **Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**
Skewness measures distribution asymmetry.

Positive skew: Tail on right, mean > median.

Negative skew: Tail on left, mean < median.

Zero skew: Symmetrical data.

Example: Income is positively skewed. Skewness impacts measure selection — median is better in skewed data. Analysts adjust for skew to ensure accurate interpretation.

# **Q7. What is the interquartile range (IQR), and how is it used to detect outliers?**
IQR = Q3 - Q1, representing the middle 50% of data.
Used to find outliers:

Lower bound: Q1 – 1.5×IQR

Upper bound: Q3 + 1.5×IQR

Values beyond these bounds are outliers.
Example: In exam scores, IQR helps detect students with exceptionally high or low marks. It measures spread and data consistency.

# **Q8. Discuss the conditions under which the binomial distribution is used.**
Binomial distribution applies when:

Fixed number of trials (n)

Two outcomes per trial (success/failure)

Constant success probability (p)

Trials are independent

Example: Tossing a coin 10 times.
It calculates the probability of getting a specific number of successes, e.g., 6 heads. It’s widely used in quality control and probability estimation.

# **Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**
Normal distribution is symmetric, bell-shaped, with mean = median = mode.
The Empirical Rule says:

68% data within 1 SD

95% within 2 SD

99.7% within 3 SD

Example: Human heights follow normal distribution. It helps in predictions and standardizing scores (z-scores) in tests and experiments.

# **Q10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.**
Example: On average, 3 calls/hour come to a call center.
Poisson applies since events occur:

Randomly

Independently

At constant rate

To find P(5 calls in 1 hour):

P(X=5) = (e^–3 × 3⁵) / 5! ≈ 0.1008

Poisson is used in traffic flow, server requests, or any rare events per interval.

# **Q11. Explain what a random variable is and differentiate between discrete and continuous random variables.**
A random variable assigns numeric values to outcomes of a random process.

Types:

Discrete: Countable outcomes (e.g., number of heads in 3 tosses).

Continuous: Infinite outcomes within a range (e.g., height, weight).

Random variables help define probability distributions like binomial or normal, crucial for data modeling and predictions.
